```python
def get_brand_name(car_name):
    car_name=car_name.split(' ')[0]
    return car_name.strip()
    
```

```python
def clean_data(value):
    value=value.split(' ')[0]
    value=value.strip()
    if value== '':
        value=0
    return float(value)
    
```

```python
get_brand_name("maruti swift dezire")
```

```python
df['name']=df['name'].apply(get_brand_name)
```

```python
df['name'].unique()
```

```python
df['mileage']=df['mileage'].apply(clean_data)
```

```python
df['max_power']=df['max_power'].apply(clean_data)
```

```python
df['engine']=df['engine'].apply(clean_data)
```

```python
brand_mapping = {
    'Maruti': 1, 'Skoda': 2, 'Honda': 3, 'Hyundai': 4, 'Toyota': 5,
    'Ford': 6, 'Renault': 7, 'Mahindra': 8, 'Tata': 9, 'Chevrolet': 10,
    'Datsun': 11, 'Jeep': 12, 'Mercedes-Benz': 13, 'Mitsubishi': 14,
    'Audi': 15, 'Volkswagen': 16, 'BMW': 17, 'Nissan': 18, 'Lexus': 19,
    'Jaguar': 20, 'Land': 21, 'MG': 22, 'Volvo': 23, 'Daewoo': 24,
    'Kia': 25, 'Fiat': 26, 'Force': 27, 'Ambassador': 28, 'Ashok': 29,
    'Isuzu': 30, 'Opel': 31
}

df['name'] = df['name'].replace(brand_mapping)
```

```python
df['transmission'].unique()
```

```python
df['transmission'] = df['transmission'].replace({
    'Manual': 1,
    'Automatic': 2
})
```

```python
df['seller_type'].unique()
```

```python
df['seller_type'] = df['seller_type'].replace({
    'Individual': 1,
    'Dealer': 2,
    'Trustmark Dealer': 3
})
```

```python
df['fuel'].unique()
```

```python
df['fuel']= df['fuel'].replace({
    'Diesel': 1,
    'Petrol': 2,
    'LPG': 3,
    'CNG': 4
})
```

```python
df.info()
```

```python
df
```

```python
df.reset_index(inplace=True)
```

```python
df['owner'].unique()
```

```python
df['owner']=   df['owner'].replace({
    'First Owner': 1,
    'Second Owner': 2,
    'Third Owner': 3,
    'Fourth & Above Owner': 4,
    'Test Drive Car':5
})
```

```python
df.drop(columns=['index'],inplace=True)
```

```python
X=df.drop(columns=['selling_price'])
y=df['selling_price']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```
